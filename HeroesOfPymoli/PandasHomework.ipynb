# Purchasing Analysis (Total)

#### - Run basic calculations to obtain number of unique items, average price, etc.
#### - Create a summary data frame to hold the results
#### - optional: give the displayed data cleaner formatting
#### - Display the summary data frame

```python
#Number of unique items.
items_df = len(purchase_data_df["Item ID"].unique())

#Avarage price calculation
average_price_df = purchase_data_df["Price"].mean()
average_price_df

#Total number of purchases.
total_purchases_df = len(purchase_data_df)
total_purchases_df


#Total Revenue.
total_revenue_df = purchase_data_df["Price"].sum()
total_revenue_df.round(2)



#Inserting result into a DAtaFrame
purchasing_analysis_df = pd.DataFrame({"Number of Unique Items":[items_df],
                                             "Average Price":[average_price_df],
                                            "Number of Purchases":[total_purchases_df],
                                             "Total Revenue":[total_revenue_df]})

purchasing_analysis_df = purchasing_analysis_df.round(2)
purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:,.2f}".format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map("${:,.2f}".format)
purchasing_analysis_df = purchasing_analysis_df.loc[:, ["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]


purchasing_analysis_df
```

# Gender Demographics

#### -Percentage and Count of Male Players
#### -Percentage and Count of Female Players
#### -Percentage and Count of Other / Non-Disclosed

```python
#Gender Count
gender_count_df = purchase_data_df.groupby("Gender")["SN"].nunique()
gender_count_df.head()

#Gender Percentage
gender_percentage_df = (gender_count_df/576)*100
gender_percentage_df.round(2)


#Gender demographics DataFrame.
gender_demographics_df = pd.DataFrame({"Total Count": gender_count_df,
                                       "Percentage of Players":gender_percentage_df})

gender_demographics_df = gender_demographics_df.round(2)

gender_demographics_df
```

# Purchasing Analysis (Gender)

#### - Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
#### -Create a summary data frame to hold the results
#### -Optional: give the displayed data cleaner formatting
#### -Display the summary data frame

```python
#Purchase count
gender_purchase_count_df= purchase_data_df.groupby("Gender")["SN"].count()
gender_purchase_count_df

#Average price of items by gender.
gender_average_df = purchase_data_df.groupby("Gender")["Price"].mean()
gender_average_df.round(2)

#Total Purchase value by gender.
gender_total_purchases_df = purchase_data_df.groupby("Gender")["Price"].sum()
gender_total_purchases_df

#Average Total purchase per person
gender_total_revenue_df = gender_total_purchases_df/gender_count_df
gender_total_revenue_df.round(2)


#Purchasing analysis DataFrame by gender.
gender_analysis_df = pd.DataFrame({"Purchase Count":gender_purchase_count_df,
                                             "Average Purchase Price":gender_average_df,
                                             "Total Purchase Value":gender_total_purchases_df,
                                             "Avg Total Purchase per Person":gender_total_revenue_df
                                          })

gender_analysis_df = gender_analysis_df.round(2)
gender_analysis_df["Average Purchase Price"] = gender_analysis_df["Average Purchase Price"].map("${:,.2f}".format)
gender_analysis_df["Total Purchase Value"] = gender_analysis_df["Total Purchase Value"].map("${:,.2f}".format)
gender_analysis_df["Avg Total Purchase per Person"] = gender_analysis_df["Avg Total Purchase per Person"].map("${:,.2f}".format)


gender_analysis_df


```

```python
#Create Bins  <10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39 >39.
bins = [0,9.99,14.99,19.99,24.99,29.99,34.99,39.99, 9999]
age_ranges = ["<10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", ">40"]
```

```python
#Cut purchase data and place the ages into bins
pd.cut(purchase_data_df["Age"], bins, labels=age_ranges)
```

```python
#Creating a group based of the bins
by_age_df = purchase_data_df.groupby("Age Range")
by_age_df.max()
```

# Purchasing Analysis (Age)

#### - Bin the purchase_data data frame by age
#### - Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
#### - Create a summary data frame to hold the results
#### - Optional: give the displayed data cleaner formatting
#### - Display the summary data frame

```python
#Purchase count by age range.
age_group_count_df = purchase_data_df.groupby("Age Range")["SN"].count()
age_group_count_df


#Purchase count by age range.
age_group_count_df = purchase_data_df.groupby("Age Range")["SN"].count()
age_group_count_df


#Average purchase price by age range.
age_group_average_df = purchase_data_df.groupby("Age Range")["Price"].mean()
age_group_average_df.round(2)


#Total purchase value by age range.
age_group_total_df = purchase_data_df.groupby("Age Range")["Price"].sum()
age_group_total_df

#Avg Total Purchase per Person
normalized_age_total_df = age_group_total_df/576
normalized_age_total_df.round(2)

# Convert to DataFrame.
age_range_df = pd.DataFrame({"Purchase Count":age_group_count_df,
                            "Average Purchase Price":age_group_average_df,
                            "Total Purchase Value": age_group_total_df,
                            "Avg Total Purchase per Person": normalized_age_total_df
})

age_range_df = age_range_df.round(2)
age_range_df["Average Purchase Price"] = age_range_df["Average Purchase Price"].map("${:,.2f}".format)
age_range_df["Total Purchase Value"] = age_range_df["Total Purchase Value"].map("${:,.2f}".format)
age_range_df["Avg Total Purchase per Person"] = age_range_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
age_range_df
```

# Top Spenders

#### - Run basic calculations to obtain the results in the table below
#### - Create a summary data frame to hold the results
#### - Sort the total purchase value column in descending order
#### - Optional: give the displayed data cleaner formatting
#### - Display a preview of the summary data frame

```python
#Purchase count by sn
sn_group_count_df = purchase_data_df.groupby("SN")["Item ID"].count()
sn_group_count_df

#Average purchase price by sn.
sn_group_average_df = purchase_data_df.groupby("SN")["Price"].mean()
sn_group_average_df.round(2)

#Total purchase value by sn.
sn_group_total_df = purchase_data_df.groupby("SN")["Price"].sum()
sn_group_total_df# Convert to DataFrame.
sn_range_df = pd.DataFrame({"Purchase Count":sn_group_count_df,
                            "Average Purchase Price":sn_group_average_df,
                            "Total Purchase Value": sn_group_total_df,
                            
})
sn_range_df = sn_range_df.round(2)
sn_range_df["Average Purchase Price"] = sn_range_df["Average Purchase Price"].map("${:,.2f}".format)
sn_range_df["Total Purchase Value"] = sn_range_df["Total Purchase Value"].map("${:,.2f}".format)

sn_range_df
```

```python
#Total items purchases analysis.

items_purchase_count_df = purchase_data_df.groupby(["Item ID", "Item Name"]).count()["Price"]
items_average_price_df = purchase_data_df.groupby(["Item ID", "Item Name"]).mean()["Price"]
items_value_total_df = purchase_data_df.groupby(["Item ID", "Item Name"]).sum()["Price"]

items_purchased_df = pd.DataFrame({"Purchase Count":items_purchase_count_df,
                                   "Item Price":items_average_price_df,
                                   "Total Purchase Value":items_value_total_df,})

items_purchased_df = items_purchased_df.round(2)
items_purchased_df["Item Price"] = items_purchased_df["Item Price"].map("${:,.2f}".format)
items_purchased_df["Total Purchase Value"] = items_purchased_df["Total Purchase Value"].map("${:,.2f}".format)

items_purchased_df.head()

```
